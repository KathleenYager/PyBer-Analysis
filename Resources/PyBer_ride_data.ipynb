```python
#Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics

# Load csv file
pyber_ride_df = pd.read_csv("PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.yticks(np.arange(0, 51, step=5.0))
plt.xticks(rotation="horizontal")
plt.color = ("skyblue")
plt.show()
```

```python
pyber_ride_df.plot.barh(x="Month", y="Avg. Fare ($USD)", color = "skyblue")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.yticks(np.arange(0, 51, step=5.0))
plt.xticks(rotation="horizontal")
plt.show()
```
